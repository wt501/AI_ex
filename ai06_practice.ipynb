```python
x = [[240, 80, 99], [60, 78, 83], [20, 32, 49], [180, 98, 85], [300, 41, 63]]

y = [432, 350, 151, 478, 401]

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x, y)

w0 = lr.intercept_
w1 = lr.coef_[0]
w2 = lr.coef_[1]
w3 = lr.coef_[2]
print(w0, w1, w2, w3) # y = w0 + w1x1 + w2x2 + w3x3

result = lr.predict([[50, 40, 80], [200, 78, 10]])
print(result)
```

# 考察

w1の値が大きいため、勉強効率が高い人ほど点数の伸びが大きいことがわかる。
