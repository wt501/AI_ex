```python
# 数値や文字列の定義および表示
x = 1
y = "apple"
txt = "私は学生です"

print(x)
print(y)
print(txt)
```

```python
# 演算子
r = (x + 5) / 2
print(r)

r = r % 2
print(r)
```

```python
# リスト
l = [12, 32, 53]
fruit = ["apple", "banana"]
v = [x, r]
n_list = list(range(10))

print(l)
print(fruit)
print(v)
print(n_list)
```

```python
# if文
x = 1
if x < 2:
    print("ok")
else:
    print("ng")
```

```python
# for文
for n in n_list:
    print(n)
```

```python
# for文とif文の組み合わせ
for n in n_list:
    print(n)
    if n < 5:
        print("ok")
    else:
        print("ng")
```

```python
# 3段ピラミッドの表示
print("  *  ")
print(" *** ")
print("*****")
```
