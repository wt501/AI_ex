```python
# データの準備
pred = [1, 1, 0, 0, 1, 0, 0, 1]
gt = [1, 0 ,0 , 0, 1, 1, 0, 0]

# 各値の定義
tp = 0
tn = 0
fp = 0
fn = 0

# 各値を計算する
for p, g in zip(pred, gt):  # predとgtからそれぞれ順に取り出す
    print(p, g)
    if p == 1 and g == 1:   # 予測が1、正解が1のとき
        tp = tp + 1         # TPに1を足す
    elif p == 1 and g == 0: # 予測が1、正解が0のとき
        fp = fp + 1         # FPに1を足す
    elif p == 0 and g == 1: # 予測が0、正解が1のとき
        fn = fn + 1         # FNに1を足す
    elif p == 0 and g == 0: # 予測が0、正解が0のとき
        tn = tn + 1         # TNに1を足す
print(tp, tn, fp, fn)

# Accuracyの計算
accuracy = (tp + tn) / (tp + tn + fp + fn)
print(accuracy)
```

```python
# MSEの計算
x = [240,  60,  20, 180, 300]
y = [432, 350, 151, 478, 401]
a = 1.0       # 直線の傾き
b = 190     # 直線の切片

fx = []                 # fxiのリストをfxとして定義
for xi in x:
    fxi = a * xi + b    # axi + bを計算する
    fx.append(fxi)      # fxにfxiを追加
print(fx)

sum = 0                 # 和の値をsumとして定義
for yi, fxi in zip(y, fx):
    sum = sum + (yi - fxi) ** 2

loss = sum / 5
print(loss)
```
