```python
x = [[240, 80, 99], [60, 78, 83], [20, 32, 49], [180, 98, 85], [300, 41, 63]]
y = [432, 350, 151, 478, 401]

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x, y)

w0 = lr.intercept_
w1 = lr.coef_[0]
w2 = lr.coef_[1]
w3 = lr.coef_[2]
print(f"y = {w0} + {w1} x1 + {w2} x2 + {w3} x3")

result = lr.predict([[50, 40, 80]])
print(result)
```
