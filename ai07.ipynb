```python
# おまじない
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler

# 学習データの準備
x = [[240, 80, 99], [60, 78, 83], [20, 32, 49], [180, 98, 85], [300, 41, 63]]
y = [432, 350, 151, 478, 401]

# 標準化
scaler = StandardScaler()
new_x = scaler.fit_transform(x)
print(new_x)
print()

# 学習
lr = LinearRegression()
lr.fit(new_x, y)

# パラメータの表示
w0 = lr.intercept_
w1 = lr.coef_[0]
w2 = lr.coef_[1]
w3 = lr.coef_[2]
print(f"y = {w0} + {w1} x1 + {w2} x2 + {w3} x3")

# 予測
y_pred = lr.predict(new_x)
print(f"prediction = {y_pred}")

# 誤差の計算
loss = mean_squared_error(y, y_pred)
print(f"loss = {loss}")

# グラフの描画
data_index = range(len(x))       # data_index = [0, 1, 2, 3, 4]
plt.scatter(data_index, y, color="red", label="Ground Truth")       # label="凡例名"
plt.scatter(data_index, y_pred, color="blue", label="Prediction")
plt.xlabel("Student")
plt.ylabel("Score")
plt.title("Train Result")
plt.xticks(data_index, ["A", "B", "C", "D", "E"]) # x軸目盛りの表示変更
plt.legend()            # 凡例の表示
plt.show()
plt.clf()               # グラフのクリア

# テストデータの準備
x_test = [[200, 50, 60], [290, 73, 71], [100, 83, 39]]
y_test = [367, 452, 388]

# 標準化
new_x_test = scaler.fit_transform(x_test)

# 予測
y_test_pred = lr.predict(new_x_test)
print(f"test prediction = {y_test_pred}")

# 誤差の計算
test_loss = mean_squared_error(y_test, y_test_pred)
print(f"test loss = {test_loss}")

# グラフの描画
data_index = range(len(x_test))       # data_index = [0, 1, 2]
plt.scatter(data_index, y_test, color="red", label="Ground Truth")
plt.scatter(data_index, y_test_pred, color="blue", label="Prediction")
plt.xlabel("Student")
plt.ylabel("Score")
plt.title("Test Result")
plt.xticks(data_index, ["test_A", "test_B", "test_C"])
plt.legend()
plt.show()
```
