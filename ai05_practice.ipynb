```python
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt

lr = LinearRegression()
x = [[240], [60]. [20], [180], [300]]
y = [432, 350, 151, 478, 401]

lr.fit(x, y)
a = lr.coef_[0]
b = lr.intercept_
print(f"{a}x + {b}")

_x = [240, 60, 20, 180, 300]# プロット用のx
lx = [0, 300]               # 始点、終点のx座標
ly = [b, a * 300 + b]       # 始点、終点のy座標
plt.plot(lx, ly, color="red")   # 直線の描画
plt.scatter(_x, y)  # 散布図の描画
plt.show()          # グラフの表示
```
