```python
# グラフを描くときに必ず必要
%matplotlib inline
import matplotlib.pyplot as plt

# データの入力
x = [240, 60,  20,  180, 300]       # データのx座標のリスト
y = [432, 350, 151, 478, 401]       # データのy座標のリスト

# 直線の描画
x1 = 0          # 始点のx座標
y1 = 200        # 始点のy座標
x2 = 300        # 終点のx座標
y2 = 500        # 終点のy座標
s = [x1, x2]    # 始点と終点のx座標のリスト
t = [y1, y2]    # 始点と終点のy座標のリスト
plt.plot(s, t, color="red", linestyle="solid")  # 直線の描画

# 直線の式の導出
a = (y2 - y1) / (x2 - x1)   # 傾きの計算
b = y1                      # 切片
print(f"{a}x + {b}")        # 直線の表示

# データの描画
plt.scatter(x, y)           # 散布図の描画
plt.title("Test Score")     # グラフタイトル
plt.xlabel("time[min]")     # x軸ラベル
plt.ylabel("score")         # y軸ラベル
plt.show()                  # グラフの表示
```
